## Notebook Set Up

```python
# # Import dependencies: PyMongo to connect to MongoDB and pprint for readable output
from pymongo import MongoClient
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# Select the 'uk_food' database and assign it to a variable
db = mongo['uk_food']
```

```python
# List all collections in the 'uk_food' database
print(db.list_collection_names())
```

```python
# Assign the 'establishments' collection to a variable for querying
establishments = db['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Query establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
print(f"Number of documents in result: {establishments.count_documents(query)}\n")

# Display the first document in the results using pprint
print("First result:")
results = establishments.find(query)
pprint(results[0])
```

```python
import pandas as pd

# Convert query results to a Pandas DataFrame
resultsHygieneDF = pd.DataFrame(results)

# Print the total number of matching records
print("Rows in DataFrame: ", len(resultsHygieneDF))

# Display the first 10 rows of the DataFrame
resultsHygieneDF.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': 'London'},
        'RatingValue': {'$gte': 4}
        }

# Count and display the number of matching documents
establishments.count_documents(query)

# Display the first document in the results using pprint
print("First result:")
results = establishments.find(query)
pprint(results[0])
```

```python
# Convert query results to a Pandas DataFrame
resultsLondonDF = pd.DataFrame(results)

# Print the total number of matching records
print("Rows in DataFrame: ", len(resultsLondonDF))

# Display the first 10 rows of the DataFrame
resultsLondonDF.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.086058
longitude = 1.196408 

query = {
    "RatingValue": 5,  # Filter establishments with a rating of 5
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
}

sort =  [("scores.Hygiene", 1)]
top5Nearby = list(establishments.find(query).sort(sort).limit(5))

# Print the results
print("Top 5 establishments near the specified location with a RatingValue of 5: \n")
for establishment in top5Nearby:
    pprint(establishment)
```

```python
# Convert query results to a Pandas DataFrame
top5DF = pd.DataFrame(top5Nearby)

# Display the DataFrame
top5DF.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Define an aggregation pipeline to analyze hygiene scores
pipeline = [
    # Step 1: Match establishments that have a Hygiene score of 0
    {"$match": {"scores.Hygiene": 0}},

    # Step 2: Group the results by Local Authority and count the number of matches per authority
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},

    # Step 3: Sort the grouped results in descending order based on the count
    {"$sort": {"count": -1}},

    # Step 4: Limit the results to the top 10 Local Authorities
    {"$limit": 10}
]

# Execute the aggregation pipeline and convert the results to a list
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the results
print(f"Number of results in pipeline: {len(results)}")

# Print the top 10 Local Authorities having establishments with a Hygiene Score of 0
print("\nTop 10 Local Authorities with the most establishments having a Hygiene Score of 0:")
pprint(results)
```

```python
# Convert pipeline results to a Pandas DataFrame
top10HygieneZeroDF = pd.DataFrame(results)

# Display the first 10 rows of the DataFrame
top10HygieneZeroDF.head(10)
```
